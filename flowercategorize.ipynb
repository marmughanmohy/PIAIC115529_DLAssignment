# labels

```python
daisylab=np.zeros((613,1))
dandlab=np.ones((842,1))
roslab=np.ones((628,1))*2
sunlab=np.ones((587,1))*3
tullab=np.ones((788,1))*4
app=np.append(daisylab,dandlab)
app1=np.append(app,roslab)
app2=np.append(app1,sunlab)
app3=np.append(app2,tullab)

#randomize
np.random.seed(77)
shuf=np.random.randint(0,len(app3),len(app3)) 
app4=app3[shuf]
display(app4)
one_hot_encode_label_train = to_categorical(app4)

display(one_hot_encode_label_train )
```

# test labels

```python
daisylab=np.zeros((769-613,1))
dandlab=np.ones((1052-842,1))
roslab=np.ones((784-628,1))*2
sunlab=np.ones((734-587,1))*3
tullab=np.ones((984-788,1))*4
app=np.append(daisylab,dandlab)
app1=np.append(app,roslab)
app2=np.append(app1,sunlab)
app3=np.append(app2,tullab)
#randomize
np.random.seed(77)
shuf=np.random.randint(0,len(app3),len(app3)) 
app5=app3[shuf]
display(app5)
one_hot_encode_label_test = to_categorical(app5)


display(one_hot_encode_label_test )
```

# trainimages append and normalize

```python
imgt=np.append(daisytrain,danditrain,axis=0)
imgt1=np.append(imgt,rosetrain,axis=0)
imgt2=np.append(imgt1,suntrain,axis=0)
totalimgtrain=np.append(imgt2,tultrain,axis=0)
```

```python
totalimgtrain.shape
```

```python
totalimgtrain= totalimgtrain / 255
pp=tultrain[782]
plt.imshow(pp)
totalimgtrain=totalimgtrain.reshape(3458,dsize[0]*dsize[1]*3)
totalimgtrain.shape
```

# randomize

```python
np.random.seed(77)
shuf=np.random.randint(0,len(totalimgtrain),len(totalimgtrain)) 
display(shuf)
totalimgtrain=totalimgtrain[shuf]
display(totalimgtrain)
```

 # testimgs append and normalize

```python
imgte=np.append(daisytest,danditest,axis=0)
imgte1=np.append(imgte,rosetest,axis=0)
imgte2=np.append(imgte1,suntest,axis=0)
totalimgtest=np.append(imgte2,tultest,axis=0)
```

```python
totalimgtest.shape
```

```python
totalimgtest = totalimgtest / 255
totalimgtest=totalimgtest.reshape(865,dsize[0]*dsize[1]*3)
```

# randomize data testimg

```python
np.random.seed(77)
shuf=np.random.randint(0,len(totalimgtest),len(totalimgtest)) 
display(shuf)
totalimgtest=totalimgtest[shuf]
display(totalimgtest)
```

# network def

```python
from keras import models
from keras import layers
network = models.Sequential()
network.add(layers.Dense(64, activation='relu', input_shape=(dsize[0] * dsize[1]*3,)))
network.add(layers.Dense(40, activation='relu'))
network.add(layers.Dense(30, activation='relu'))
network.add(layers.Dense(5, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
x_val = totalimgtrain[:1800]
partial_x_train = totalimgtrain[1800:]
y_val = one_hot_encode_label_train[:1800]
partial_y_train = one_hot_encode_label_train[1800:]
```

```python
history = network.fit(partial_x_train,partial_y_train,epochs=200,batch_size=100,validation_data=(x_val, y_val))
```

```python
network.evaluate(totalimgtest,one_hot_encode_label_test)
```
